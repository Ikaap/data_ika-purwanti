# 1. Buatlah sebuah data visualization dari dataset [berikut](https://gist.github.com/nadirbslmh/7a3d729818dd09bb195ef998033638ce) dengan kriteria:
- Membuat sebuah line plot untuk menggambarkan jumlah nilai dari transaksi per hari.

```python
import pandas as pd
import matplotlib.pyplot as plt

df_transaction = pd.read_csv(r"D:\Alterra Academy\tugas\data_ika-purwanti\20_Data-Visualization\praktikum\source\electronic_shop.csv")
df_transaction.set_index("transaction_id", inplace=True)

transaction_sum_day = df_transaction.groupby(["transaction_date"]).sum()

transaction_dates = transaction_sum_day.index.get_level_values(0).tolist()
total_amount = transaction_sum_day["transaction_amount"]

plt.title("Total Amount Transactions per Day")
plt.xlabel("Transaction Dates")
plt.ylabel("Total Transaction Amount")

plt.plot(transaction_dates, total_amount, marker="o")
plt.grid(True)
plt.show()
```

- Membuat sebuah bar plot untuk menggambarkan jumlah customer per hari.

```python
transaction_count_day = df_transaction.groupby(["transaction_date"]).count()

transaction_dates = transaction_count_day.index.get_level_values(0).tolist()
customer_count_day = transaction_count_day["customer_name"]

plt.title("Total Customer per Day")
plt.xlabel("Transaction Dates")
plt.ylabel("Total Customer")

plt.bar(transaction_dates, customer_count_day)
plt.show()
```

- Membuat sebuah scatter plot untuk menggambarkan jumlah nilai dari transaksi per hari.

```python
plt.title("Total Amount Transactions per Day")
plt.xlabel("Transaction Dates")
plt.ylabel("Total Transaction Amount")

plt.scatter(transaction_dates, total_amount)
plt.show()
```

# 2. Sebuah toko roti / bakery telah merekap [data transaksi](https://gist.github.com/nadirbslmh/17b1bcb77023d7d030bb7d39968d45f1). Toko tersebut ingin mendapatkan informasi berikut:
- Jumlah nilai transaksi per hari.

```python
df_transaction_bakery = pd.read_csv(r"D:\Alterra Academy\tugas\data_ika-purwanti\20_Data-Visualization\praktikum\source\bakery.csv")
df_transaction_bakery.set_index("transaction_id", inplace=True)

transaction_sum_day_bakery = df_transaction_bakery.groupby(["transaction_date"]).sum()

transaction_dates = transaction_sum_day_bakery.index.get_level_values(0).tolist()
total_amount = transaction_sum_day_bakery["transaction_amount"]

plt.figure(figsize=(10, 6))
plt.title("Total Amount Transactions per Day")
plt.xlabel("Transaction Dates")
plt.ylabel("Total Transaction Amount")

plt.plot(transaction_dates, total_amount, marker="o")
plt.xticks(rotation=60)
plt.grid(True)
plt.show()
```

- Jenis roti yang paling banyak dipesan.

```python
menu_bread_result = df_transaction_bakery.groupby(["item_name"]).count()

bread_name = menu_bread_result.index.get_level_values(0).tolist()
bread_count = menu_bread_result["transaction_date"]

plt.figure(figsize=(10, 10))
plt.pie(
    bread_count, 
    labels=bread_name,
    autopct="%1.1f%%",
    startangle=90,
)
plt.title("Most Ordered Bread Type")
plt.show()
```

- Rata-rata kepuasan pelanggan (customer_rating) per hari.

```python
transaction_date_result = df_transaction_bakery.groupby(["transaction_date"]).mean(numeric_only=True)

transaction_dates = transaction_date_result.index.get_level_values(0).tolist()
amount_avg = transaction_date_result["customer_rating"]

plt.figure(figsize=(10, 6))
plt.title("Customer Rating Average per Day")
plt.xlabel("Transaction Dates")
plt.ylabel("Averages")

plt.plot(transaction_dates, amount_avg, marker="o")
plt.xticks(rotation=60)
plt.grid(True)
plt.show()
```
